## Utility Function(s) & Variable(s)

```python
# Function to write extracted text output into txt file
def write_txt(extracted_info, package_used):
    filename = package_used + '_textfile.txt' 
    textfile = open(filename, 'w')
    textfile.write(extracted_info)
    textfile.close()
```

## PyPDF2 - via URL (worst performance)

```python
import requests
import PyPDF2
import io
```

```python
def extract_content(url):
    # retrieve PDF binary stream
    response = requests.get(url)
    open_pdf_file = io.BytesIO(response.content)
    pdf = PyPDF2.PdfFileReader(open_pdf_file)

    # return concatenated content
    text = [pdf.getPage(i).extractText() for i in range(0, pdf.getNumPages())]
    return "\n".join(text)
```

```python
pypdf2_content=extract_content(url_2)
write_txt(pypdf2_content, 'pypdf2')
```

## PDFMiner - via URL

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO, BytesIO
import urllib 
```

```python
def pdf_from_url_to_txt(url):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    f = urllib.request.urlopen(url).read()
    fp = BytesIO(f)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos = set()
    for page in PDFPage.get_pages(fp,
                                  pagenos,
                                  maxpages=maxpages,
                                  password=password,
                                  caching=caching,
                                  check_extractable=False):
        interpreter.process_page(page)
    fp.close()
    device.close()
    str = retstr.getvalue()
    retstr.close()
    return str
```

```python
pdfminer_content = pdf_from_url_to_txt(url_2)
write_txt(pdfminer_content, 'pdfminer')
```

Observations 
- a lot of \n even within paragraphs (would be hard to differentiate between different paragraphs)
- good for mini paragraphs (pg 7, 11)
- unable to extract diagrams' information
- most text are extracted

```python
print(pdfminer_content)
```

## PDFMiner.six via URL

```python
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser
```

```python
def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

## PDFStructure - via URL (best recommended - with headers identified)

```python
from pdfstructure.hierarchy.parser import HierarchyParser
from pdfstructure.source import FileSource
from io import StringIO, BytesIO
import urllib 
```

```python
pdfstructure_pdf = urllib.request.urlopen(url_1).read()
pdfstructure_pdf = BytesIO(pdfstructure_pdf)
```

```python
parser = HierarchyParser() 

# specify source (that implements source.read())
source = FileSource(pdfstructure_pdf) 
# analyse document and parse as nested data structure
document = parser.parse_pdf(source)
```

```python
from pdfstructure.printer import PrettyStringPrinter

pretty_string_printer = PrettyStringPrinter()
pretty_string = pretty_string_printer.print(document)
print(pretty_string)
```

## Data Preprocessing

```python
import re
```

```python
# remove and replace all (cid: XXX) info in pretty_string
cleaned_string = re.sub('\(cid\:[0-9][0-9]+\)', '', pretty_string)
```

```python
# print(cleaned_string)
```

```python
# find all headers available for extraction of relevant sections using regex 
# source = https://stackoverflow.com/questions/4894069/regular-expression-to-return-text-between-parenthesis
# \[(.*?)\]
all_headers = re.findall('\[.*?\]',cleaned_string)
all_headers
```

```python
len(all_headers)
```

```python
# method to filter for relevant decarbonisation sections 
def decarbon_filter(all_headers):
    decarbon_headers = []
    decarbon_terms = {'energy', 'emission', 'emissions', 'renewable', 'carbon', 'fuel', 'power', 'green', 'gas', 'green energy', 
                         'sustainable', 'climate', 'sustainability', 'environmental', 'environment', 'GHG', 'decarbon'}
    for header in all_headers:
        header_set = set(header[1:-1].lower().split())
        if header_set.intersection(decarbon_terms):
            decarbon_headers.append(header)
    return decarbon_headers
            
decarbon_headers = decarbon_filter(all_headers)
decarbon_headers
```

```python
print(len(decarbon_headers))
```

```python
# (?<=ABC).*?(?=XYZ)
# decarbon_regex = []
# for header in decarbon_headers:
#     regex = header[1:-1] + '(.*?)' + '\['
#     decarbon_regex.append(regex)
```

```python
# section to compile all relevant decarbonisation content
decarbon_content = ''
for header in decarbon_headers:
    start = pretty_string.find(header) + len(header)
    nxt_idx = all_headers.index(header) + 1
    if nxt_idx < len(all_headers):
        nxt_header = all_headers[nxt_idx]
        end = pretty_string.find(nxt_header)
        content = pretty_string[start:end]
        decarbon_content += (header + content)
```
